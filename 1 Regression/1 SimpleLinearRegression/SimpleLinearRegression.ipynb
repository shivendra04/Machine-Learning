```python
import pandas as pd# to import dataset and create matrices
import matplotlib.pyplot as plt# to drauw chart
import numpy as np# to work with array
```

```python
#Importing teh dataset
```

```python
dataset=pd.read_csv('C:\\Users\\BizAct-110\\OneDrive\\ML_Practice\\Regression\\SimpleLinearRegression\\Salary_Data.csv')
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
#Spliting the dataset into Training set and test set
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
#Trainig the simple regression model on the training set
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)# it will trian the model on training set
```

```python
#Predicting the Training set result
y_pred=regressor.predict(x_test)
```

```python
#Vizualizing the Training set Result
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,regressor.predict(x_train),color='blue')
plt.title('Salary Vs Experience(Training set)')
plt.xlabel('Year of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
#Vizualizing the Test set Result
plt.scatter(x_test,y_test,color='red')
plt.plot(x_train,regressor.predict(x_train),color='blue')
plt.title('Salary Vs Experience(Test set)')
plt.xlabel('Year of Experience')
plt.ylabel('Salary')
plt.show()
```
